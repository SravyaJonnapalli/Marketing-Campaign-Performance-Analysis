```python
import pandas as pd
df=pd.read_csv(r"C:\Users\User\Downloads\marketing_campaign_dataset.csv\marketing_campaign_dataset.csv")
```

```python
df.head()
```

```python
import re

# Gender
df['Gender'] = df['Target_Audience'].str.lower().apply(
    lambda x: 'Female' if 'women' in x 
    else ('Male' if 'men' in x else 'All')
)

# Age group
df['Age_Group'] = df['Target_Audience'].apply(
    lambda x: re.search(r'\d{2}-\d{2}', x).group() 
    if re.search(r'\d{2}-\d{2}', x) else 'All Ages'
)
```

```python
df.head()
```

```python
df['Duration_Days'] = df['Duration'].str.replace(' days', '', regex=False).astype(int)
```

```python
df.head()
```

```python
df['Acquisition_Cost'] = (
    df['Acquisition_Cost']
    .astype(str)
    .str.replace('$', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)
```

```python
df.head()
```

```python
df['Acquisition_Cost'].dtype
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='mixed', dayfirst=True)
```

```python
df.info()
```

```python
# CTR
df['click_through_rate'] = df['Clicks'] / df['Impressions']

# CPC
df['cost_per_click'] = df['Acquisition_Cost'] / df['Clicks']

# CPM
df['cost_per_impression'] = (df['Acquisition_Cost'] / df['Impressions']) * 1000

# Profit
df['Profit'] = df['ROI'] * df['Acquisition_Cost']
```

```python
df.head()
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Month_Name'] = df['Date'].dt.month_name()
```
